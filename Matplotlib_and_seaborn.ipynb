# Styling lines and markers

The plt.plot function supports many arguments for styling lines and markers:<br>
color or c: set the color of the line (supported colors)<br>
linestyle or ls: choose between a solid or dashed line<br>
linewidth or lw: set the width of a line<br>
markersize or ms: set the size of markers<br>
markeredgecolor or mec: set the edge color for markers<br>
markeredgewidth or mew: set the edge width for marker <br>
markerfacecolor or mfc : set the fill color for markers<br>
alpha : opacity of the plot


```python
plt.plot(years,apples,marker="o",c='b',ls='--',lw=3,ms=6,mew=5,mec='navy');
plt.plot(years,oranges,marker="x",c='r',ls='-',lw=6,ms=6,mew=3,mec='black',alpha=.5);

plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)');
plt.title('Crop of yield in India')
# for know which colour plot is for which fruit
plt.legend(['Apples','Oranges']);
```

```python
#using fmt dirctlywe can write s-r
plt.plot(years,apples,'s-b');
plt.plot(years,oranges,'x--r');

plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)');
plt.title('Crop of yield in India')
# for know which colour plot is for which fruit
plt.legend(['Apples','Oranges']);
```
